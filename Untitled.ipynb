```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
match.shape
```

```python
delivery.head()
```

```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df.head()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
pip install sklearn
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(f"Training Set : {X_train.shape}\nTesting Set : {X_test.shape}")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
models=dict()
```

# Logistic Regression


```python
LogReg=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
LogReg.fit(X_train,y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
train_score_Logreg = str(LogReg.score(X_train, y_train) * 100)
test_score_Logreg = str(LogReg.score(X_test, y_test) * 100)
print(f'Train Score : {train_score_Logreg[:5]}%\nTest Score : {test_score_Logreg[:5]}%')
models["LogReg"] = test_score_Logreg
```

# Random Forest Classifier

```python
Rf=Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
```

```python
Rf.fit(X_train,y_train)
```

```python
y_pred = Rf.predict(X_test)
```

```python
train_score_Rf = str(Rf.score(X_train, y_train) * 100)
test_score_Rf = str(Rf.score(X_test, y_test) * 100)
print(f'Train Score : {train_score_Rf[:5]}%\nTest Score : {test_score_Rf[:5]}%')
models["Rf"] = test_score_Rf
```

# Support vector classification

```python
svc=Pipeline(steps=[
    ('step1',trf),
    ('step2',SVC(C=10,probability=True))
])     
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
train_score_svc = str(svc.score(X_train, y_train) * 100)
test_score_svc = str(svc.score(X_test, y_test) * 100)
print(f'Train Score : {train_score_svc[:5]}%\nTest Score : {test_score_svc[:5]}%')
models["svc"] = test_score_svc
```

```python
import matplotlib.pyplot as plt 
model_names = list(models.keys())
accuracy = list(map(float, models.values()))
# creating the bar plot
plt.title("Comparision of Algorithms")
plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.bar(model_names, accuracy)
```
